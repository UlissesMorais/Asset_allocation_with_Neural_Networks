## Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Boxplots

### FIA1

```python
NN_FIA1_relu_df = pd.read_csv(r'./predictions_performace/FIA1_3x20layers_relu_20142022_NN_performance.csv')
NN_FIA1_sig_df = pd.read_csv(r'./predictions_performace/FIA1_3x20layers_sigmoid_20142022_NN_performance.csv')
NN_FIA1_relu_df1 = pd.read_csv(r'./predictions_performace/FIA1_6x20layers_relu_20142022_NN_performance.csv')
NN_FIA1_sig_df1 = pd.read_csv(r'./predictions_performace/FIA1_6x20layers_sigmoid_20142022_NN_performance.csv')

MKW_FIA1_df = pd.read_csv(r'./predictions_performace/FIA1_MKW_performance.csv')

# plot boxplots
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(16, 12))


axs[0, 0].boxplot([NN_FIA1_relu_df['2019'], NN_FIA1_relu_df['2020'], NN_FIA1_relu_df['2021'], NN_FIA1_relu_df['2022']])
axs[0, 0].set_title('FIA1 - Neural Nets 3x20 layers & relu')
axs[0, 0].set_ylabel('Neural Sharpes > Random Sharpes')

axs[0, 1].boxplot([NN_FIA1_sig_df['2019'], NN_FIA1_sig_df['2020'], NN_FIA1_sig_df['2021'], NN_FIA1_sig_df['2022']])
axs[0, 1].set_title('FIA 1 - Neural Nets 3x20 layers & sigmoid')

axs[0, 2].boxplot([MKW_FIA1_df['2019'], MKW_FIA1_df['2020'], MKW_FIA1_df['2021'], MKW_FIA1_df['2022']])
axs[0, 2].set_title('FIA1 - Markowitz')
axs[0, 2].set_ylabel('Markowitz Sharpes > Random Sharpes')


axs[1, 0].boxplot([NN_FIA1_relu_df1['2019'], NN_FIA1_relu_df1['2020'], NN_FIA1_relu_df1['2021'], NN_FIA1_relu_df1['2022']])
axs[1, 0].set_title('FIA 1 - Neural Nets 6x20 layers & relu')
axs[1, 0].set_ylabel('Neural Sharpes > Random Sharpes')

axs[1, 1].boxplot([NN_FIA1_sig_df1['2019'], NN_FIA1_sig_df1['2020'], NN_FIA1_sig_df1['2021'], NN_FIA1_sig_df1['2022']])
axs[1, 1].set_title('FIA 1 - Neural Nets 6x20 layers & sigmoid')

axs[1, 2].boxplot([MKW_FIA1_df['2019'], MKW_FIA1_df['2020'], MKW_FIA1_df['2021'], MKW_FIA1_df['2022']])
axs[1, 2].set_title('FIA1 - Markowitz')
axs[1, 2].set_ylabel('Markowitz Sharpes > Random Sharpes')


for ax in axs.flatten():
    
    # ax.yaxis.grid(True)
    ax.set_yticks(np.arange(0, 1.1, 0.1))
    ax.axhline(y = 0.5, color = 'g', linestyle = '--')
    ax.set_xticks([1, 2, 3, 4])
    ax.set_xticklabels(['2019','2020','2021','2022'])
    ax.set_xlabel('periods')
    # ax.xlabel_outer()

plt.show()
```
